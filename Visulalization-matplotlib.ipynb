
# Bar Plot

```python
plt.xlabel('chalo')
plt.ylabel('chalta hai')
plt.bar(x1,y1,label="apple",)
plt.bar(x2,y2,label="MICOSOFT")
plt.plot(x2,y2,label="amazon",c='green')
plt.grid(c='red')
plt.legend()
```
